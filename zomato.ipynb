```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
df = pd.read_csv(r"F:\Downloads\Zomato data .csv")
print(df)
```

```python
df
```

```python
def handleRate(val):
    val= str(val).split('/')
    val=val[0]
    return float(val)

df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
print(df.head())
```

```python
import seaborn as sns


sns.countplot(x=df['listed_in(type)'])
plt.xlabel("type of resturant")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  # ✅ This line is required!

sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Restaurant")       # Optional: avoids overlap
plt.show()               # ✅ Displays the graph
```

```python
groupData = df.groupby('listed_in(type)')['votes'].sum()
res = pd.DataFrame({'votes': groupData})

plt.plot(res, c="green", marker="o")  # corrected 'marker' argument
plt.xlabel('type of res', c="red", size=20)
plt.ylabel('votes', c="red", size=20)
plt.show()  # Don't forget to show the plot
```

```python
plt.hist(df['rate'],bins=5)
plt.title('rating')
plt.show()
```

```python
coupleData=df['approx_cost(for two people)']
sns.countplot(x=coupleData ,color="green")
```

```python
plt.Figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate' , data=df)
```

```python
pivot=df.pivot_table(index='listed_in(type)' , columns='online_order' , aggfunc='size',fill_value=0)
sns.heatmap(pivot,annot=True,cmap='YlGnBu',fmt='d')
plt.title('heatmap')
plt.xlabel('online order')
plt.ylabel('list_in(type)')
plt.show()
```
