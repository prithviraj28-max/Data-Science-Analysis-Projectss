```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"F:\Downloads\compressed_data.csv.gz")
```

```python
df.columns
```

```python
print(df.isnull().sum())
```

Changing the data type

```python
df["last review"] = pd.to_datetime(df["last review"], errors='coerce')
```

```python
df.info()
```

Replace null value with zero   ,  inplace is used to change permanently

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace=True)
```

Remove all the null permanently 

```python
df.dropna(subset=['NAME', 'host name'], inplace=True)
```

Delete columns 

```python
df= df.drop(columns=['license','house_rule'], errors='ignore')
```

REmove symbol and Make cahnge its data type

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

Remove duplicate

```python
df.drop_duplicates(inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded your data
# df = pd.read_csv('your_airbnb_data.csv')

# Convert 'last review' column to datetime
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Group by 'last review' date (by month) and count the number of reviews
reviews_per_date = df.groupby(df['last review'].dt.to_period('M')).size()

# Convert PeriodIndex to datetime for plotting
reviews_per_date.index = reviews_per_date.index.to_timestamp()

# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(reviews_per_date.index, reviews_per_date.values, color='red', linewidth=2)
plt.title('How has the number of reviews changed over time?')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```
